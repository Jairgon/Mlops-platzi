```python
import os
import json
import pandas as pd
```

```python
def read_json(path: str, file_name: str):
    """This method is used to read the json file
    Args:
        path (str): path of the file
        file_name (str): name of the file
    Returns:
        df_tickets (pd.DataFrame): dataframe with the tickets"""
    file_path = os.path.join(path, file_name)
    with open(file_path, "r") as file:
        datos = json.load(file)
    df_tickets = pd.json_normalize(datos)
    return df_tickets
```

```python
df_tickets = read_json(path = "data/data_raw", 
                       file_name = "tickets_classification_eng.json")
```
